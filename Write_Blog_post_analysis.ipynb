```python
# to handle datasets
import numpy as np
import pandas as pd
from collections import Counter

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)

# to make plots
import matplotlib.pyplot as plt
import seaborn as sns

# to change plot style
sns.set(style="darkgrid")

# to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading Stackoverflow survey data
df_2017 = pd.read_csv('survey_results_public_2017.csv')
df_2018 = pd.read_csv('survey_results_public_2018.csv')
df_2019 = pd.read_csv('survey_results_public_2019.csv')
```

```python
def filter_country(df, column_filter, country, columns):
    ''' Filter a dataframe by country.,
        returns filtred dataframe.
    '''
    
    df_copy = df
    
    for column in columns: 
        df_copy = df_copy[df_copy[column_filter] == country].dropna(subset=[column])
        
    return df_copy

# Filter the dataframe by India and selecting some columns.
India_2017 = filter_country(df_2017, 'Country', 'India', ['HaveWorkedLanguage', 'WantWorkLanguage'])
India_2017.rename({'HaveWorkedLanguage':'LanguageWorkedWith','WantWorkLanguage':'LanguageDesireNextYear'}, axis=1, inplace=True)
India_2018 = filter_country(df_2018, 'Country', 'India', ['LanguageWorkedWith', 'LanguageDesireNextYear'])
India_2019 = filter_country(df_2019, 'Country', 'India', ['LanguageWorkedWith', 'LanguageDesireNextYear'])
India_2019

# DB Level.
DB_India_2017 = filter_country(df_2017, 'Country', 'India', ['HaveWorkedDatabase', 'WantWorkDatabase'])
DB_India_2017.rename({'HaveWorkedDatabase':'DatabaseWorkedWith','WantWorkDatabase':'DatabaseDesireNextYear'}, axis=1, inplace=True)
DB_India_2018 = filter_country(df_2018, 'Country', 'India', ['DatabaseWorkedWith', 'DatabaseDesireNextYear'])
DB_India_2019 = filter_country(df_2019, 'Country', 'India', ['DatabaseWorkedWith', 'DatabaseDesireNextYear'])
DB_India_2019

# PF Level.
PF_India_2017 = filter_country(df_2017, 'Country', 'India', ['HaveWorkedPlatform', 'WantWorkPlatform'])
PF_India_2017.rename({'HaveWorkedPlatform':'PlatformWorkedWith','WantWorkPlatform':'PlatformDesireNextYear'}, axis=1, inplace=True)
PF_India_2018 = filter_country(df_2018, 'Country', 'India', ['PlatformWorkedWith', 'PlatformDesireNextYear'])
PF_India_2019 = filter_country(df_2019, 'Country', 'India', ['PlatformWorkedWith', 'PlatformDesireNextYear'])
PF_India_2019
```

```python
def split_column(df, column):
    ''' Split column by ;,
        returns a splited series.
    '''
    
    df_copy = df
    column_series = df_copy[column].apply(lambda x: x.split(';'))
    
    return column_series

# Splitting the Data Frame by column into a Series. 
worked_lang_2017 = split_column(India_2017, 'LanguageWorkedWith')
wanted_lang_2017 = split_column(India_2017, 'LanguageDesireNextYear')

worked_lang_2018 = split_column(India_2018, 'LanguageWorkedWith')
wanted_lang_2018 = split_column(India_2018, 'LanguageDesireNextYear')

worked_lang_2019 = split_column(India_2019, 'LanguageWorkedWith')
wanted_lang_2019 = split_column(India_2019, 'LanguageDesireNextYear')

worked_lang_overall = worked_lang_2017.append(worked_lang_2018)
worked_lang_overall = worked_lang_overall.append(worked_lang_2019)

wanted_lang_overall = wanted_lang_2017.append(wanted_lang_2018)
wanted_lang_overall = wanted_lang_overall.append(wanted_lang_2019)

#DB Level
worked_db_2017 = split_column(DB_India_2017, 'DatabaseWorkedWith')
wanted_db_2017 = split_column(DB_India_2017, 'DatabaseDesireNextYear')

worked_db_2018 = split_column(DB_India_2018, 'DatabaseWorkedWith')
wanted_db_2018 = split_column(DB_India_2018, 'DatabaseDesireNextYear')

worked_db_2019 = split_column(DB_India_2019, 'DatabaseWorkedWith')
wanted_db_2019 = split_column(DB_India_2019, 'DatabaseDesireNextYear')

worked_db_overall = worked_db_2017.append(worked_db_2018)
worked_db_overall = worked_db_overall.append(worked_db_2019)

wanted_db_overall = wanted_db_2017.append(wanted_db_2018)
wanted_db_overall = wanted_db_overall.append(wanted_db_2019)

#PF Level
worked_pf_2017 = split_column(PF_India_2017, 'PlatformWorkedWith')
wanted_pf_2017 = split_column(PF_India_2017, 'PlatformDesireNextYear')

worked_pf_2018 = split_column(PF_India_2018, 'PlatformWorkedWith')
wanted_pf_2018 = split_column(PF_India_2018, 'PlatformDesireNextYear')

worked_pf_2019 = split_column(PF_India_2019, 'PlatformWorkedWith')
wanted_pf_2019 = split_column(PF_India_2019, 'PlatformDesireNextYear')

worked_pf_overall = worked_pf_2017.append(worked_pf_2018)
worked_pf_overall = worked_pf_overall.append(worked_pf_2019)

wanted_pf_overall = wanted_pf_2017.append(wanted_pf_2018)
wanted_pf_overall = wanted_pf_overall.append(wanted_pf_2019)
```

```python
def disarray(array_list):
    ''' Flat a nested list,
        returns a flat list.
    '''
    
    objects = []
    
    for row in array_list:
        for obj in row:
            objects.append(obj.strip())
            
    return objects

# Flatting nested list objects.
list_worked_languages_2017 = disarray(worked_lang_2017)
list_wanted_languages_2017 = disarray(wanted_lang_2017)

list_worked_languages_2018 = disarray(worked_lang_2018)
list_wanted_languages_2018 = disarray(wanted_lang_2018)

list_worked_languages_2019 = disarray(worked_lang_2019)
list_wanted_languages_2019 = disarray(wanted_lang_2019)

#DB Level
list_worked_db_2017 = disarray(worked_db_2017)
list_wanted_db_2017 = disarray(wanted_db_2017)

list_worked_db_2018 = disarray(worked_db_2018)
list_wanted_db_2018 = disarray(wanted_db_2018)

list_worked_db_2019 = disarray(worked_db_2019)
list_wanted_db_2019 = disarray(wanted_db_2019)

#PF Level
list_worked_pf_2017 = disarray(worked_pf_2017)
list_wanted_pf_2017 = disarray(wanted_pf_2017)

list_worked_pf_2018 = disarray(worked_pf_2018)
list_wanted_pf_2018 = disarray(wanted_pf_2018)

list_worked_pf_2019 = disarray(worked_pf_2019)
list_wanted_pf_2019 = disarray(wanted_pf_2019)
```

```python
def group_list(data_list, year):
    ''' Group by count to a list,
        returns a result dict.
    '''
    
    grouped_list = dict(Counter(data_list))
    grouped_dict = [{'Programming Language':key, 'Count': value, 'Year': year} for key, value in grouped_list.items()]
    
    return grouped_dict

# Groping a list and creating a dict.
dict_worked_languages_2017 = group_list(list_worked_languages_2017, '2017')
dict_wanted_languages_2017 = group_list(list_wanted_languages_2017, '2017')

dict_worked_languages_2018 = group_list(list_worked_languages_2018, '2018')
dict_wanted_languages_2018 = group_list(list_wanted_languages_2018, '2018')

dict_worked_languages_2019 = group_list(list_worked_languages_2019, '2019')
dict_wanted_languages_2019 = group_list(list_wanted_languages_2019, '2019')

#DB Level

def group_list(data_list, year):
    ''' Group by count to a list,
        returns a result dict.
    '''
    
    grouped_list = dict(Counter(data_list))
    grouped_dict = [{'Databases':key, 'Count': value, 'Year': year} for key, value in grouped_list.items()]
    
    return grouped_dict

# Groping a list and creating a dict.
dict_worked_db_2017 = group_list(list_worked_db_2017, '2017')
dict_wanted_db_2017 = group_list(list_wanted_db_2017, '2017')

dict_worked_db_2018 = group_list(list_worked_db_2018, '2018')
dict_wanted_db_2018 = group_list(list_wanted_db_2018, '2018')

dict_worked_db_2019 = group_list(list_worked_db_2019, '2019')
dict_wanted_db_2019 = group_list(list_wanted_db_2019, '2019')

#PF Level

def group_list(data_list, year):
    ''' Group by count to a list,
        returns a result dict.
    '''
    
    grouped_list = dict(Counter(data_list))
    grouped_dict = [{'Platform':key, 'Count': value, 'Year': year} for key, value in grouped_list.items()]
    
    return grouped_dict

# Groping a list and creating a dict.
dict_worked_pf_2017 = group_list(list_worked_pf_2017, '2017')
dict_wanted_pf_2017 = group_list(list_wanted_pf_2017, '2017')

dict_worked_pf_2018 = group_list(list_worked_pf_2018, '2018')
dict_wanted_pf_2018 = group_list(list_wanted_pf_2018, '2018')

dict_worked_pf_2019 = group_list(list_worked_pf_2019, '2019')
dict_wanted_pf_2019 = group_list(list_wanted_pf_2019, '2019')
```

```python
def create_dataframe(data_dicts):
    ''' Create two dataframes and append them,
        returns a appended dataframe.
    '''
        
    df1 = pd.DataFrame(data_dicts[0])
    df2 = pd.DataFrame(data_dicts[1])
    df3 = pd.DataFrame(data_dicts[2])

    df = df1.append(df2)
    df = df.append(df3)
    
    return df

# Creating a Data Frame appending the data.
worked_languages = create_dataframe([dict_worked_languages_2017,dict_worked_languages_2018, dict_worked_languages_2019])
wanted_languages = create_dataframe([dict_wanted_languages_2017,dict_wanted_languages_2018, dict_wanted_languages_2019])
worked_languages

#DB Level
worked_db = create_dataframe([dict_worked_db_2017,dict_worked_db_2018, dict_worked_db_2019])
wanted_db = create_dataframe([dict_wanted_db_2017,dict_wanted_db_2018, dict_wanted_db_2019])
worked_db

#PF Level
worked_pf = create_dataframe([dict_worked_pf_2017,dict_worked_pf_2018, dict_worked_pf_2019])
wanted_pf = create_dataframe([dict_wanted_pf_2017,dict_wanted_pf_2018, dict_wanted_pf_2019])
worked_pf
```

```python
def percentage(df, column):
    ''' Scale data,
        returns data scaled.
    '''
    
    df_copy = df
    series = []
    
    for val in df_copy[column].unique():
        series.append(df_copy[df_copy[column] == val]['Count'] / 
                      df_copy[df_copy[column] == val]['Count'].sum())
    
    joined_series = pd.Series()
    for i_series in series:
        joined_series = joined_series.append(i_series)
        
    return joined_series

# Calculating the percentage
worked_languages['Percentage'] = percentage(worked_languages, 'Year')
wanted_languages['Percentage'] = percentage(wanted_languages, 'Year')

#DB Level
worked_db['Percentage'] = percentage(worked_db, 'Year')
wanted_db['Percentage'] = percentage(wanted_db, 'Year')

#PF Level
worked_pf['Percentage'] = percentage(worked_pf, 'Year')
wanted_pf['Percentage'] = percentage(wanted_pf, 'Year')
```

```python
# Get the top 10 languages
top_10_worked_lang = worked_languages.sort_values(by=['Percentage'], ascending=False).head(10)['Programming Language'].unique()
top_10_wanted_lang = wanted_languages.sort_values(by=['Percentage'], ascending=False).head(10)['Programming Language'].unique()

worked_lang_chart = worked_languages[worked_languages['Programming Language'].isin(top_10_worked_lang)]
wanted_lang_chart = wanted_languages[wanted_languages['Programming Language'].isin(top_10_worked_lang)]

#DB Level
top_10_worked_db = worked_db.sort_values(by=['Percentage'], ascending=False).head(10)['Databases'].unique()
top_10_wanted_db = wanted_db.sort_values(by=['Percentage'], ascending=False).head(10)['Databases'].unique()

worked_db_chart = worked_db[worked_db['Databases'].isin(top_10_worked_db)]
wanted_db_chart = wanted_db[wanted_db['Databases'].isin(top_10_worked_db)]

#PF Level
top_10_worked_pf = worked_pf.sort_values(by=['Percentage'], ascending=False).head(10)['Platform'].unique()
top_10_wanted_pf = wanted_pf.sort_values(by=['Percentage'], ascending=False).head(10)['Platform'].unique()

worked_pf_chart = worked_pf[worked_pf['Platform'].isin(top_10_worked_pf)]
wanted_pf_chart = wanted_pf[wanted_pf['Platform'].isin(top_10_worked_pf)]
```

```python
# Grouping data to create a new dataframe

rows = []
for itens in list(zip(worked_lang_overall, wanted_lang_overall)):
    for i in itens[0]:
        rows.append({
            'Worked_PL': i, 
            'Wanted_PL': Counter(itens[0])
        })
        
programming_language_transition = pd.DataFrame(rows).groupby('Worked_PL').agg({'Wanted_PL': 'sum'}).reset_index()

#DB Level
rows = []
for itens in list(zip(worked_db_overall, wanted_db_overall)):
    for i in itens[0]:
        rows.append({
            'Worked_DB': i, 
            'Wanted_DB': Counter(itens[0])
        })
        
DB_transition = pd.DataFrame(rows).groupby('Worked_DB').agg({'Wanted_DB': 'sum'}).reset_index()

#PF Level
rows = []
for itens in list(zip(worked_pf_overall, wanted_pf_overall)):
    for i in itens[0]:
        rows.append({
            'Worked_PF': i, 
            'Wanted_PF': Counter(itens[0])
        })
        
PF_transition = pd.DataFrame(rows).groupby('Worked_PF').agg({'Wanted_PF': 'sum'}).reset_index()
```

```python
# Creating a matrix

for row in programming_language_transition['Worked_PL']:
    programming_language_transition[row] = 0

for index, row in programming_language_transition.iterrows():
    total = sum([value for key, value in dict(row['Wanted_PL']).items()])
    for key, value in dict(row['Wanted_PL']).items():
         programming_language_transition[key].loc[index] = (value / total)
            
#DB Level
for row in DB_transition['Worked_DB']:
    DB_transition[row] = 0

for index, row in DB_transition.iterrows():
    total = sum([value for key, value in dict(row['Wanted_DB']).items()])
    for key, value in dict(row['Wanted_DB']).items():
         DB_transition[key].loc[index] = (value / total)
            
#PF Level
for row in PF_transition['Worked_PF']:
    PF_transition[row] = 0

for index, row in PF_transition.iterrows():
    total = sum([value for key, value in dict(row['Wanted_PF']).items()])
    for key, value in dict(row['Wanted_PF']).items():
         PF_transition[key].loc[index] = (value / total)
```

```python
## Figure 7 — Percentage of relationship of work programming languages and desire programming 
## languages in 2017, 2018 and 2019.

matrix = programming_language_transition.drop('Wanted_PL', axis=1)\
                                        .set_index('Worked_PL')

plt.figure(figsize=(16, 10))
sns.heatmap(matrix, cmap="YlGnBu")
plt.title('Programming Languages Heatmap', size = 16)
plt.ylabel('')
plt.show()
```

```python
## Figure 8 — Percentage of relationship of work Databases and desire Databases in 2017, 2018 and 2019.

matrix = DB_transition.drop('Wanted_DB', axis=1)\
                                        .set_index('Worked_DB')

plt.figure(figsize=(16, 10))
sns.heatmap(matrix, cmap="YlGnBu")
plt.title('Databases Heatmap', size = 16)
plt.ylabel('')
plt.show()
```

```python
## Figure 9 — Percentage of relationship of work Platforms and desire Platforms in 2017, 2018 and 2019.

matrix = PF_transition.drop('Wanted_PF', axis=1)\
                                        .set_index('Worked_PF')

plt.figure(figsize=(16, 10))
sns.heatmap(matrix, cmap="YlGnBu")
plt.title('Platforms Heatmap', size = 16)
plt.ylabel('')
plt.show()
```
